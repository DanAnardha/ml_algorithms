```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
df = pd.read_excel('baseball2.xlsx')
df
```

```python
label_encoder = LabelEncoder()
df_encoded = df.apply(label_encoder.fit_transform)
df_encoded
```

```python
X = df_encoded[['Outlook', 'Temp', 'Humidity', 'Wind']]
y = df_encoded['Play']
```

```python
clf = DecisionTreeClassifier(criterion='entropy')
clf.fit(X, y)
```

```python
plt.figure(figsize=(7, 7))
tree.plot_tree(clf, rounded=True, filled=True, feature_names=X.columns, class_names=df[df.columns[-1]].values)
plt.show()
```

```python
tree_rules = export_text(clf, feature_names=X.columns)
print(tree_rules)
```

```python
predicted = clf.predict(X)
predicted_labels = label_encoder.inverse_transform(predicted)
print(df[df.columns[-1]].values)
print("Hasil Prediksi:\n", predicted_labels)
```

```python
print(classification_report(df[df.columns[-1]].values, predicted_labels))
```
